```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_data(N, M, T, target_sharpe):
    beta = np.random.randn(M, 1)
    beta = np.sqrt(target_sharpe ** 2 / 12) * beta / np.sqrt(np.sum(beta ** 2)) / np.sqrt(N)

    signals = np.random.randn(N, M, 2 * T)
    epsilon = np.random.randn(N, 2 * T)
    returns = np.tensordot(signals, beta, axes=(1, 0)).squeeze() + epsilon
    in_sample_signals = signals[:, :, :T]
    in_sample_returns = returns[:, :T]

    out_of_sample_signals = signals[:, :, T:]
    out_of_sample_returns = returns[:, T:]
    return beta, in_sample_signals, out_of_sample_signals, in_sample_returns, out_of_sample_returns
```

```python
def performance_of_strategy(returns, signals, betas):
    number_signals = signals.shape[1]

    # signals_t  = N \times p

    # factor_returns(t, k) = \sum_{i=1}^N x_{i,t}(k) R_{i,t+1} , k=1,...,p
    factor_returns = np.sum(signals.transpose((1, 0, 2))
                            * np.tile(returns, (number_signals, 1, 1)), 1)
    returns = np.matmul(betas.T, factor_returns)
    std_ret = returns.std()
    mean_ret = returns.mean()
    sharpe = np.round(np.sqrt(12) * mean_ret / std_ret, 2)
    return sharpe, returns
```

```python
if __name__ == '__main__':

    N = 1000
    p = 100
    T = 120  # 10 years
    for sample in range(100):
        # generate data
        beta, in_sample_signals, out_of_sample_signals, in_sample_returns, out_of_sample_returns \
            = generate_data(N, p, T, target_sharpe=1)

        # run OLS
        # this is X'X
        bar_s = np.tensordot(in_sample_signals.transpose((1, 0, 2)), in_sample_signals, axes=[(1, 2), (0, 2)])

        lam = 0.0
        bar_s = bar_s + lam * np.identity(p) # lambda is the ridge parameter

        factor_means = np.tensordot(in_sample_signals.transpose((1, 0, 2)),
                                    in_sample_returns).reshape(-1, 1)  # this is X'y

        # this is the OLS estimate of beta
        beta_ols = np.matmul(np.linalg.inv(bar_s), factor_means)

        # compute strategy returns in sample
        is_sharpe, is_returns = performance_of_strategy(in_sample_returns, in_sample_signals,
                                                        beta_ols)

        oos_sharpe, oos_returns = performance_of_strategy(out_of_sample_returns, out_of_sample_signals,
                                                          beta_ols)

        fig, axes = plt.subplots(1, 2)
        axes[0].plot(is_returns.cumsum())
        axes[0].title.set_text(f'In sample, SR={is_sharpe}')
        axes[1].plot(oos_returns.cumsum())
        axes[1].title.set_text(f'Out of sample, SR={oos_sharpe}')
        plt.show()
```
